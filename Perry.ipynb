```python
import pandas as pd
WHO_df = pd.read_csv("WHO Covid Data(in).csv")

WHO_df.head()
```

```python
for i in range(len(WHO_df)):
    if i == 0:
        WHO_df.loc[i, 'Percent_increase_cases'] = 0
    
    if WHO_df.loc[i, 'New_cases'] == 0:
        WHO_df.loc[i, 'Percent_increase_cases'] = 0 

    else:
        WHO_df.loc[i, 'Percent_increase_cases'] = (WHO_df.loc[i,'New_cases'] - WHO_df.loc[i-1, 'New_cases']) / WHO_df.loc[i,'New_cases']


```

```python
WHO_df.iloc[2500:]
```

```python
plt.plot(WHO_df.loc[:, 'Date_reported'], WHO_df.loc[:, 'Percent_increase_cases'])

WHO_df['Date_reported'] = pd.to_datetime(WHO_df['Date_reported'])
plt.show()
```
